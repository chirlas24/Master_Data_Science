````python
npoints = 10000
mu = 50; sigma = 20;
X2 = np.random.randn(npoints)*sigma + mu
````

````python
from scipy.stats import norm
dist = norm(mu, sigma)
n = 50
xnorm = np.linspace(mu-3*sigma, mu+3*sigma, n)
ynorm = dist.pdf(xnorm)
plt.figure(figsize=(10,10))
plt.plot(xnorm, ynorm, 'g', label=r'$\mu=%s, \sigma =%s$' % (mu, sigma))
plt.hist(X2, bins=100, density=True, label='Histogram')
plt.legend()
plt.xlabel('my dist', fontsize=15)
plt.grid(False)
````

# Fill between

````python
Y1 = X
Y2 = X/2
plt.plot(X, Y1, 'r', label='Y1')
plt.plot(X, Y2, 'b', label='Y2')
plt.legend()
plt.fill_between(X,X,X/2, where=Y1>2, color='g', alpha=0.4)
plt.fill_between(X,X,X/2, where=Y1<-2, color='y', alpha=0.4)
plt.xlim([-7,7]);
````

### Ejercicio

````python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2*X)

plt.figure(figsize=(10,10))
plt.plot(X, Y+1, color='blue', alpha=1)
plt.plot(X, Y-1, color='blue', alpha=1)
plt.grid(False)
plt.fill_between(X,Y+1,1, where=Y+1>1, color='g', alpha=0.4)
plt.fill_between(X,Y+1,1, where=Y+1<1, color='peru', alpha=0.4)
plt.fill_between(X,Y-1,-1, where=Y+1>1, color='orange', alpha=0.4)
plt.fill_between(X,Y-1,-1, where=Y+1<1, color='r', alpha=0.4)
plt.xticks([])
plt.yticks([]);
````

# Ejercicio:

Crea una figura que muestre la función $\sin(x)$ y $\exp(x)$, entre $-\pi$ y $\pi$, con diferentes estilos para cada función (lineas con puntos en azul y lineas discontinuas en rojo). La figura debe incluir leyenda explicativa y nombres de ejes:
* xlabel = '$\alpha$ (rad)'
* ylabel = 'f'

Sin ningún tipo de grid.
Por último, la figura debe sombrear el área entre las dos curvas.

````python
import math
````

````python
X = np.linspace(-math.pi, math.pi, 50)
Y1 = np.sin(X)
Y2 = np.exp(X)

figure, axes = plt.subplots(1,1)

axes.plot(X, Y1,'b.', label=r'$\alpha (rad)$')
axes.plot(X, Y2,'r--', label='f')
axes.set_xlim(-math.pi, math.pi)
axes.grid(None)
axes.legend('best', shadow=True)
axes.set_title('EJERCICIO PLOT')
axes.fill_between(X,Y1,Y2, color='g', alpha=0.2)
````
