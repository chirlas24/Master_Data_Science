```python
import numpy as np
def deptime_to_string(num):
    if num!=num: #Comprobando si es NaN
        num =0.0;

    hour = int(num / 100) % 24 
    minute = int(num % 100)
    

    return '%02d:%02d' % (hour, minute)

deptime_to_string(2423.0)
```

```python
series = df2['DepTime'].dropna().apply(deptime_to_string)
series.sample(5)
```
