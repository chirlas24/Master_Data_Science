```python
#Function to find the header value

def header_value(file_path, sheet):
    header = 0
    for row in range(0,51):
        aux_file = pd.read_excel(file_path, sheet_name=sheet, header=header)
        if aux_file.index[0] == 'DISTRITO_ACCIDENTE':
            return header
    raise ValueError ('Header cannot be more than 50')
```
