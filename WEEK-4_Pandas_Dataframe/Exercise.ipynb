```python
import pandas as pd
import numpy as np
```

```python
def create_column_year(dataframe, year):
    dataframe['YEAR'] = year
    return dataframe
```

```python
def open_file(file_path, year):
    return pd.read_excel(file_path, 
                                index_col=0,
                                header=7, 
                                sheet=year, 
                                skip_footer=1,
                                usecols='A:K')
```

```python
open_file(file_path, 2009)
```

```python
file_path = '207831-0-accidentes-trafico.xls'
```

```python
df_total = open_file(file_path, 2009)
df_total = create_column_year(df_total, 2009)
df_total
```

```python
for year in range(2010,2017):
    df_aux = open_file(file_path, year)
    df_aux = create_column_year(df_aux, year)
    df_total = pd.concat([df_total, df_aux])
```
