### Plotting

```python
#Cambiamos al fichero de verdad para que el plot tenga sentido, si no seria todo el mismo mes.
searches = pd.read_csv('searches.csv.bz2', delimiter='^', usecols=['Date', 'Destination'], nrows=100000)
searches = searches.dropna()
searches = searches[searches['Destination'].isin(['MAD', 'BCN', 'AGP'])]
searches['Year-Month'] = searches['Date'].str[0:7]
del(searches['Date'])
searches['Searches'] = 1
searches_grouped = searches.groupby(['Destination', 'Year-Month'])
searches_grouped = searches_grouped.count().reset_index()
```

```python
searches_grouped.head()
```

```python
A = searches_grouped.pivot('Year-Month', 'Destination', 'Searches')
```

```python
print(A)
```

```python
A.plot(figsize=(10,10))
```

```python
#Lo que llevamos de codigo
searches = pd.read_csv('searches.csv.bz2', delimiter='^', usecols=['Date', 'Destination'], nrows=100000)
searches = searches.dropna()
searches = searches[searches['Destination'].isin(['MAD', 'BCN', 'AGP'])]
searches['Year-Month'] = searches['Date'].str[0:7]
del(searches['Date'])
searches['Searches'] = 1
searches_grouped = searches.groupby(['Destination', 'Year-Month'])
searches_grouped = searches_grouped.count().reset_index()
A = searches_grouped.pivot('Year-Month', 'Destination', 'Searches')
A.plot(figsize=(10,10))
```

## Implement in the whole document (BIGDATA)

```python
searches0 = pd.read_csv('searches.csv.bz2', delimiter='^', usecols=['Date', 'Destination'], chunksize=500000)


sum_all_chunks = pd.DataFrame()

for i, chunk in enumerate(searches0):
    
    searches = chunk.dropna()
    searches = searches[searches['Destination'].isin(['MAD', 'BCN', 'AGP'])]
    searches['Year-Month'] = searches['Date'].str[0:7]
    del(searches['Date'])
    searches['Searches'] = 1
    searches_grouped = searches.groupby(['Destination', 'Year-Month'])
    searches_grouped = searches_grouped.count().reset_index()
    sum_all_chunks = sum_all_chunks.append(searches_grouped)
        
sum_all_chunks = sum_all_chunks.groupby(['Destination', 'Year-Month'])       
sum_all_chunks = sum_all_chunks.sum().reset_index()
A = sum_all_chunks.pivot('Year-Month', 'Destination', 'Searches')
A.plot(figsize=(10,10))
```
